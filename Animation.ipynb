#Animation

```python
import matplotlib.pyplot as plt
import matplotlib.animation as mpla
import numpy as np

T = np.linspace(0, 2*np.pi, 100)
S = np.sin(T)
line = plt.plot(T,S)

def animate(i):
  line.set.ydata(np.sin(T+i/25))
anim = mpla.FuncAnimation(plt.gcf(), animate, interval=5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(i):
    line.set_ydata(np.sin(x + i / 50))  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, interval=20, blit=True, save_count=50)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(i):
     line.set_ydata(np.sin(x + i / 10)) # update the data.
     return line,


ani = animation.FuncAnimation(
     fig, animate, interval=20, blit=True, save_count=100)

ani.save("movie.mp4") # download file and run


plt.show()
```

#Linear regression

```python
#data input
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2.2, 4.1, 6.7, 8.0, 9.8, 13]

#esteemed model
import numpy as np
mod_linear = np.polyfit(x, y, 1)
new_y = np.polyval(mod_linear, x)

# building graph
import matplotlib.pyplot as plt
plt.plot(x, y, "o", label="experimental points")
plt.plot(x, new_y, "r-", label="model")
plt.xlabel("x")
plt.xlabel("y")
plt.title("fitting a straight line")
plt.legend()
plt.show()
```

https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

numpy.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)

Returns: pndarray, shape (deg + 1,) or (deg + 1, K) Polynomial coefficients, highest power first. If y was 2-D, the coefficients for k-th data set are in p[:,k].

Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y). Returns a vector of coefficients p that minimises the squared error in the order deg, deg-1,  0.

```python
#data input
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2.2, 4.1, 6.7, 8.0, 9.8, 13]

  #esteemed model
import numpy as np
mod_linear = np.polyfit(x, y, 1)
new_y = np.polyval(mod_linear, x)
print(mod_linear)

  # building graph

import matplotlib.pyplot as plt
plt.plot(x, y, "o", label="experimental points")
plt.plot(x, new_y, "r-", label="model")
plt.xlabel("x")
plt.xlabel("y")
plt.title("fitting a straight line")
plt.legend()
plt.show()

# calculate coefficient of determination
y_resid = y - new_y
SQresid = sum(y_resid**2)
y_tot = y - np.mean(y)
SQtotal = sum(y_tot**2)
R2= 1 - SQresid/SQtotal
print(f"The adjustment determination coefficient is {R2}")
```
